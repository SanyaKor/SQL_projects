# Задание 1

Необходимо вывести названия треков, которые были приобретены компаниями, которые обслуживаются сотрудниками, нанятыми после 1 сентября 2002 года. 
    

```python
%load_ext sql
%sql sqlite:///chinook.db
```

```python
%%sql 
select distinct t.name 
from invoice_items ii
join invoices i
on ii.invoiceid = i.invoiceid 
join customers c
on i.customerid = c.customerid 
join employees e
on c.supportrepid = e.employeeid 
join tracks t 
on ii.trackid = t.trackid 
where datetime(e.hiredate)>'2002-09-01 0:0:0'
```

```python
%%sql 
select p.name,p.playlistid,sum(unitprice) as total_price,(sum(milliseconds)/60000) as total_time
from tracks t 
join playlist_track pt
on t.trackid = pt.trackid
join playlists p
on p.playlistid = pt.playlistid
group by pt.playlistid,p.name
```

## Задание 3
Для каждого артиста определить все жанры песен, которые у него есть и количество песен по каждому жанру.


```python
%%sql 
select a.name as artist,g.name as genre,count(trackid)as amount
from artists a
join albums ab
on a.artistid = ab.artistid
join tracks t
on t.albumid = ab.albumid
join genres g
on t.genreid = g.genreid
group by a.artistid,g.genreid,g.name,a.name
```

## Задание 4
Для каждого артиста найти жанр, по которому он заработал наибольшее количество денег, а также вывести соотношение общей суммы, заработанной артистом по всем другим жанрам к общей сумме заработанных денег.

```python
%%sql 
with 
    AA
    as ( 
         select ar.artistid, ar.name artist, g.name genre, sum(ii.unitprice * ii.quantity) as total_price 
            from invoice_items ii 
                 join tracks t
                 on ii.trackid = t.trackid 
                 join albums a
                 on t.albumid = a.albumid 
                 join artists ar
                 on a.artistid = ar.artistid 
                 join genres g 
                 on t.genreid = g.genreid 
                 group by ar.artistid, g.genreid, ar.name, g.name
        ), 
    PP 
    as ( 
         select AA.artistid, max(AA.total_price) as best_price,
        (sum(AA.total_price) - max(AA.total_price)) / sum(AA.total_price) koef
             from AA 
                 group by artistid) 
select AA.artist Artist, AA.genre Genre, PP.koef
from AA, PP  
where AA.total_price = PP.best_price 
and AA.artistid = PP.artistid
group by Artist,genre,koef
order by PP.artistid
```

## Задание 5
Для каждой компании вывести общее количестов песен, купленных по жанру рок

```python
%%sql
select c.company,sum(ii.quantity) as  kolvo
from customers c
join invoices i
on i.customerid = c.customerid
join invoice_items ii
on i.invoiceid = ii.invoiceid
join tracks t
on t.trackid = ii.trackid
join genres g
on t.genreid = g.genreid
where g.name like 'Rock'
group by c.company
```

## Задание 6
Вывести альбом, который является минимальным по объему(в байтах) среди альбомов со стоимостью, большей средней цены за альбом по всем альбомам с таким же жанром (если у альбома песни с нескольким жанрами, то жанром является тот, у которого наибольшее количество песен в альбоме)

```python
%%sql
with AlbumsGeneres
as
(
    select albumid,Title,genreid as RealGenre,max(HM)
    from
         (select A.Title,A.albumid,G.genreid,G.Name,count(G.genreid) as HM
        from
            tracks T
            join albums A using(albumid)
            join genres G using (genreid)
            group by A.albumid,G.genreid
         )
    group by Title
),
AVG_by_Real_Genre
as
    (   
        select avg(fullPrice) as AAvg, RealGenre
            from(
                select RealGenre,sum(unitprice) as fullPrice,A.Title
                    from AlbumsGeneres
                    join albums A using(albumid)
                    join tracks T using (albumid)
                    group by albumid
    )
    group by RealGenre
),
AlbumSize_Fprice
as
    (
        select sum(unitprice) as Totalprice,sum(bytes) as size,albumid
        from Tracks T
        join Albums A using(albumid)
        group by albumid
    ),
RealMax
as
(
    select max(size) as MaxSize
        from AlbumsGeneres
        join AVG_by_Real_Genre AAVG using(RealGenre)
        join AlbumSize_Fprice ASF using(albumid)
        where TotalPrice>AAvg
)

 select Title, max(size) as Size
        from AlbumsGeneres
        join AVG_by_Real_Genre AAVG using(RealGenre)
        join AlbumSize_Fprice ASF using(albumid)
        join RealMax RM on RM.MaxSize=ASF.size 
        where TotalPrice>AAvg
        group by RealGenre,Title,Maxsize
        having size=MaxSize
```
